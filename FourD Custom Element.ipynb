# FourD Custom Element
FourD is a 3D force directed gaph layout written in javascript. It depends on threejs, which enables it to display 3D graphics in the browser. 

Recent developments in HTML allow for the creation of custom elements to be included in the familiar DOM. 

Creating a custom fourd element will take some learning, but the result should be an immensly usable experience. Anyone with HTML skills will be able to include a graph on their page. 

## Class Design
In the following sections, I will go over the details of FourD's class design. 

The basic operations that make up a dynamic graph are:

* addition of a vertex
* removal of a vertex
* addition of an edge
* removal of an edge

It maintains the following state: 

* list of vertices
* list of edges

where vertices are of class Vertex, comprising: 

* an id unique to the graph
* display instructions, such as
  * color or picture
  * size
* display properties, which are generated by the implementation, such as
  * position
  * velocity
  * acceleration
* a possibly empty list of edges

and edges are of class Edge, containing: 

* a reference to the source vertex
* a reference to the target vertex
* a boolean indicating whether the edge is directed (displayed down pointing). 
* undefined, as of yet, a strength fraction with which to manipulate the length of the edge. 

The original Ubigraph configuration even had a way to specify vertex and edge attributes, but that's not supported by fourd, yet.

```
* ```js

```
* ```js 
vertex = fourd.graph.add_vertex(...)
```
* ```js 
fourd.graph.remove_vettex(vertex)
```

* ```js
edge = fourd.graph.add_edge(vertex1, vertex2)
```

* ```js 
fourd.gaph.remove_edge(edge)
```

I think we can do without something called Graph in our final HTML, but a fourd element will contain a set of edges and vertices.
